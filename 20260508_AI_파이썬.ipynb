```python
#리스트 컴프리헨션: 반복이나 조건문 넣어서 리스트 값을 짧게 쭉 이어쓰게 해주는건데 간결하긴하나 풀어써도 되는걸 묶어서 쓰니 적는게 복잡해져서 딱히 장점이 없
nums = [1, 2, 3, 4, 5]
print([n*n for n in nums])
res = []
res += ["even"] if 4 % 2 == 0 else ["odd"]
print(res)
```

```python
double = lambda x: x * 2  #임의함수 일괄적용
print(double(5))

str_nums = ["1", "2", "3", "10"]  #그냥 map을 이용한 일괄적용
nums = list(map(int, str_nums))
print(nums)
```

```python
nums = [1, 2, 3, 4, 5, 6, 7, 8, 9]

odds = list(filter(lambda x: x % 2 == 1, nums))   #필터로 거르는건데 람다 넣어서 일괄 적용해 걸러냄
print(odds)
```
