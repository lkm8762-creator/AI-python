```python
#점수 입출력 변환장치
score = int(input('점수를 입력해주세요.'))
if score >100:
  print('점수는 100점이 최대입니다.')
elif score >= 90:
  print(score,'점으로 A입니다.')
elif 90 > score >= 80:
  print(score,'점으로 B입니다.')
elif 80 > score >= 70:
  print(score,'점으로 C입니다.')
elif 70 > score >= 60:
  print(score,'점으로 D입니다.')
else:
  print(score,'점으로 F입니다.')
```

#for문들, 지정 시작~끝+1까지 반복, range엔 1씩 증감이 붙어있음

```python
for i in range (1, 101):
    print(i)
```

```python
#짝수만 반복
for i in range(0, 101, 2):
    print(i)
```

```python
#합 출력
sum=0
for i in range(1, 101, 1):
    sum = sum+i
    print(sum)
```

```python
#짝홀 각각 더하기
sum1=0
sum2=0
for i in range(1, 101):
    if i%2==0:
        sum1=sum1+i
    else:
        sum2=sum2+i
print('짝수합:',sum1)
print('홀수합:',sum2)
```

```python
#숫자 맞추기 게임-교수님ver
import random
answer=random.randint(0,1000)  #맞추고자 하는 답은 난수로 생성
count=0   #횟수라는 새로운 값을 얻기 위해 새 변수를 하나 추가
user=int(input('생각한 정답을 적어주세요.'))   #첫 입력
while user != answer:
  if user < answer:
    print('더 큰 수를 입력하시오')
    count+=1
  else:
    print('더 작은 수를 입력하시오')
    count+=1
  user=int(input('추측값을 입력하시오'))  #틀렸을때 반복하여 받을 입력
print(f'총 {count}번 만에 맞췄습니다!')
```

```python
#약수 구하기(범위만 잡아주면 반복이 명확하니 for문)
for num in range(0,100,1):
  divisor=[]
  for i in range(1, 1+num) :
    if num % i == 0:
     divisor.append(i)
print(divisor)
```
