```python
#용돈 주사위 도박게임-10만원 이상 뽑는 경우 찾기
import random
i=0  #횟수 계산용 변수
while True:  #무한반복문
  num1=random.randint(1,6)  #매번 달라지는 주사위 숫자들
  num2=random.randint(1,6)
  num3=random.randint(1,6)
  i+=1
  print(num1, num2, num3)  #중간 결과 확인
  if num1*num2*num3*1000 >= 100000:  #최종 승인 기준
    break  #정지어
print(f'총 {i}번 실행-({num1}, {num2}, {num3})', num1*num2*num3*1000,'원입니다.')
```

```python
# =========================
# Python List: 메서드 + 내장함수 예시 모음 (Colab 복붙용)-산지니가 해준거
# =========================

# -------------------------
# 0) 기본 리스트 준비
# -------------------------
nums = [3, 1, 2, 1]
words = ["banana", "apple", "cherry"]

print(nums, words)

# =========================
# 1) 리스트 메서드 (list methods)
# =========================

# 1-1) append(x): 맨 뒤에 1개 추가
a = [1, 2]
a.append(3)
print(a)  # [1, 2, 3]

# 1-2) extend(iterable): 여러 개를 한 번에 추가(이어붙이기)
a = [1, 2]
a.extend([3, 4, 5])
print(a)  # [1, 2, 3, 4, 5]

# 1-3) insert(i, x): i 위치에 삽입
a = [1, 3, 4]
a.insert(1, 2)  # index 1 자리에 2 삽입
print(a)  # [1, 2, 3, 4]

# 1-4) remove(x): 값 x인 첫 요소 삭제 (없으면 에러)
a = [1, 2, 2, 3]
a.remove(2)  # 첫 번째 2만 삭제
print(a)  # [1, 2, 3]

# 1-5) pop([i]): i의 요소를 꺼내며 삭제 (기본은 맨 뒤를 터뜨림)
a = [10, 20, 30]
last = a.pop()      # 마지막 꺼내기
first = a.pop(0)    # 0번째 꺼내기
print(last, first, a)  # 30 10 [20]

# 1-6) clear(): 전부 삭제
a = [1, 2, 3]
a.clear()
print(a)  # []

# 1-7) index(x[, start[, end]]): x의 첫 위치 (없으면 에러)
a = [5, 6, 7, 6]
print(a.index(6))        # 1
print(a.index(6, 2))     # 3  # start=2부터 찾기

# 1-8) count(x): x 개수
a = [1, 1, 2, 3, 1]
print(a.count(1))  # 3

# 1-9) sort(key=None, reverse=False): 제자리 정렬(원본이 바뀜)
a = [3, 1, 2]
a.sort()
print(a)  # [1, 2, 3]

# 1-10) sort with key: 길이 기준 정렬 등
a = ["aaa", "b", "cc"]
a.sort(key=len)  # 문자열 길이 기준 오름차순
print(a)  # ['b', 'cc', 'aaa']

# 1-11) reverse(): 제자리 뒤집기
a = [1, 2, 3]
a.reverse()
print(a)  # [3, 2, 1]

# 1-12) copy(): 얕은 복사(shallow copy)
a = [1, 2, 3]
b = a.copy()
b.append(4)
print(a, b)  # [1, 2, 3] [1, 2, 3, 4]

# =========================
# 2) 리스트에 자주 쓰는 내장함수 (builtins)
# =========================

# 2-1) len(): 길이
a = [10, 20, 30]
print(len(a))  # 3

# 2-2) min(), max(): 최솟값/최댓값
a = [3, 1, 2]
print(min(a), max(a))  # 1 3

# 2-3) sum(): 합 (숫자 리스트)
a = [1, 2, 3, 4]
print(sum(a))      # 10
print(sum(a, 100)) # 110  # start=100부터 더하기

# 2-4) sorted(): 정렬된 "새 리스트" 반환(원본 유지)
a = [3, 1, 2]
b = sorted(a)
print(a, b)  # [3, 1, 2] [1, 2, 3]

# 2-5) reversed(): 역순 이터레이터(보통 list()로 감싸서 사용)
a = [1, 2, 3]
print(list(reversed(a)))  # [3, 2, 1]

# 2-6) enumerate(): (인덱스, 값)으로 반복
a = ["A", "B", "C"]
for i, v in enumerate(a):
    print(i, v)

# 2-7) any(), all(): 조건 검사(리스트 원소들의 truthy/falsey)
a = [0, 0, 1]
print(any(a))  # True  # 하나라도 참이면 True
print(all(a))  # False # 모두 참이어야 True

# 2-8) list(): 반복가능(iterable) -> 리스트 변환
print(list("abc"))  # ['a', 'b', 'c']
print(list((1, 2))) # [1, 2]

# 2-9) tuple(), set(): 다른 자료형으로 변환
a = [1, 1, 2, 3]
print(tuple(a))  # (1, 1, 2, 3)
print(set(a))    # {1, 2, 3}  # 중복 제거(순서 개념은 약함)

# 2-10) map(): 함수 적용 결과(이터레이터) -> 보통 list(map(...))
a = [1, 2, 3]
print(list(map(lambda x: x * 10, a)))  # [10, 20, 30]

# 2-11) filter(): 조건 통과한 것만(이터레이터) -> 보통 list(filter(...))
a = [1, 2, 3, 4, 5]
print(list(filter(lambda x: x % 2 == 0, a)))  # [2, 4]

# 2-12) zip(): 여러 리스트를 묶어서 튜플로
names = ["kim", "lee", "park"]
scores = [90, 80, 100]
print(list(zip(names, scores)))  # [('kim', 90), ('lee', 80), ('park', 100)]

# 2-13) dict(): (key, value) 쌍 리스트 -> dict
pairs = [("a", 1), ("b", 2)]
print(dict(pairs))  # {'a': 1, 'b': 2}

# =========================
# 3) 리스트 문법/연산(함수/메서드는 아니지만 자주 씀)
# =========================

# 3-1) 인덱싱 / 슬라이싱
a = [10, 20, 30, 40, 50]
print(a[0], a[-1])    # 10 50
print(a[1:4])         # [20, 30, 40]
print(a[::-1])        # [50, 40, 30, 20, 10]

# 3-2) 연결(+), 반복(*)
a = [1, 2]
b = [3, 4]
print(a + b)   # [1, 2, 3, 4]
print(a * 3)   # [1, 2, 1, 2, 1, 2]

# 3-3) 포함 여부(in)
a = ["apple", "banana"]
print("apple" in a)  # True
print("kiwi" in a)   # False

# 3-4) del로 삭제(인덱스/슬라이스)
a = [0, 1, 2, 3, 4, 5]
del a[0]
print(a)  # [1, 2, 3, 4, 5]
del a[1:3]  # index 1~2 삭제
print(a)  # [1, 4, 5]

# =========================
# 4) 자주 헷갈리는 포인트 한 번에 확인
# =========================

# 4-1) sort()는 원본을 바꾸고 None을 반환
a = [3, 2, 1]
ret = a.sort()
print(a, ret)  # [1, 2, 3] None

# 4-2) sorted()는 원본 유지 + 새 리스트 반환
a = [3, 2, 1]
b = sorted(a)
print(a, b)  # [3, 2, 1] [1, 2, 3]
```

```python
fruits = ['apple', 'banana', 'cherry', 'orange', 'grape', 'mango', 'pineapple',
          'strawberry', 'blueberry', 'peach', 'pear', 'kiwi', 'watermelon']

# 1) 과일 이름을 모두 대문자로 바꿔서 "한번에" 출력
upper_fruits = [f.upper() for f in fruits]
print("대문자 변환:", upper_fruits)

# 2) 과일 이름 길이를 기준으로 정렬 (짧은 것 -> 긴 것)
sorted_by_len = sorted(fruits, key=len)
print("길이 기준 정렬:", sorted_by_len)

# (원하면 긴 것 -> 짧은 것)
sorted_by_len_desc = sorted(fruits, key=len, reverse=True)
print("길이 기준 내림차순:", sorted_by_len_desc)

# 3) 이름 길이가 6글자 이상인 것만 모아서 다른 리스트로 출력
len_6_or_more = [f for f in fruits if len(f) >= 6]
print("6글자 이상:", len_6_or_more)
```

```python
fruits = ['apple', 'banana', 'cherry', 'orange', 'grape', 'mango', 'pineapple',
          'strawberry', 'blueberry', 'peach', 'pear', 'kiwi', 'watermelon']

# 1) 0번부터 번호 붙여 출력
for i, fruit in enumerate(fruits):
    i+=1    #start 없이 임의로 1부터 시작하게 조정해주는 것-이거 안하면 컴퓨터순서대로 0부터 출발함
    print(i, fruit)

print("-" * 20)

# 2) 1번부터 번호 붙여 출력
for i, fruit in enumerate(fruits, start=1):   #start로 지정해서 준것
    print(f"{i}. {fruit}")
```
