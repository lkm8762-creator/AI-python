```python
#교수님 버전
height1=float(input("학생1의 키(cm)를 적어주세요"))
height2=float(input("학생2의 키(cm)를 적어주세요"))
if height1>height2:
  print('첫번째 학생이 더 크다')
else:
  print('두 번째 학생이 더 크다')   #둘이 키가 같은 경우에 결과가 잘못나오게 됨/ 이다 아니다 정도 이분법적 구분만 되기 때문
```

```python
#교수님 버전(elif 없이 if 노가다)-권장하지 않는 형태(bs.전부 따로 실행됨, 한방에 볼걸 하나하나 다 보는 격)
height1=float(input("학생1의 키(cm)를 적어주세요"))
height2=float(input("학생2의 키(cm)를 적어주세요"))
if height1>height2:
  print('첫번째 학생이 더 크다')
if height1 == height2:
  print('두 학생은 키가 같다')
if height1 < height2:
  print('두 번째 학생이 더 크다')
```

```python
#교수님 버전(elif)-elif는 몇 개든 추가 가능함
height1=float(input("학생1의 키(cm)를 적어주세요"))
height2=float(input("학생2의 키(cm)를 적어주세요"))
if height1>height2:
  print('첫번째 학생이 더 크다')
elif height1 == height2:
  print('두 학생은 키가 같다')
else:
  print('두 번째 학생이 더 크다')
```

```python
#학점 판독기
sc=int(input("점수를 입력하세요: "))   #정수로 된 점수 입력
if sc>=95:      #if문과 elif문을 이용해 5점 단위로 학점을 구분
  print("A+")    #서로 간의 범위가 겹치지 않게 비교연산자를 잘 구분하여 써야함
elif sc >= 90:
  print("A0")
elif sc >= 85:
  print("B+")
elif sc >= 80:
  print("B0")
elif sc >= 75:
  print("C+")
elif sc >= 70:
  print("C0")
else:
  print("D")
print(sc)
```

```python
#가위바위보(1가위 2바위 3보)
import random
man1= random.randint(1,3)
man2= random.randint(1,3)
                #or로 복잡하게 쓰면 조건문 2개로 됨/ 비긴 경우(if) 사람1이 이긴 경우(elif) 사람2가 이긴 경우(else)
if man1 == man2:    # 1=1 2=2 3=3인 경우를 모두 담아냄
  if man1 == 1:     #비긴 경우 두 남자가 어떤 것을 내어 비겼는지 알기 위해 작성된 중첩if문 코드
    print('man1=가위 man2=가위')
  elif man1 == 2:
    print('man1=바위 man2=바위')
  else:
    print('man1=보 man2=보')   #이까지 중첩문
  print('두 사람이 비겼습니다.')
elif man1 == 1 and man2 == 2:
  print('두 번째 사람이 이겼습니다.','man1=가위 man2=바위')
elif man1 == 1 and man2 == 3:
  print('첫 번째 사람이 이겼습니다.','man1=가위 man2=보')
elif man1 == 2 and man2 == 1:
  print('첫 번째 사람이 이겼습니다.','man1=바위 man2=가위')
elif man1 == 2 and man2 == 3:
  print('두 번째 사람이 이겼습니다.','man1=바위 man2=보')
elif man1 == 3 and man2 == 1:
  print('두 번째 사람이 이겼습니다.','man1=보 man2=가위')
else:      #else는 조건식이 아님, 걍 짜투리/ if와 elif만 조건식으로 봄
  print('첫 번째 사람이 이겼습니다.','man1=보 man2=바위')
```
