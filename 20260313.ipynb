```python
# 사칙연산(+, -, *, /)을 하는 아주 간단한 프로그램
# (변수 + 연산자 + 출력문만 사용)

a = 12          # 첫 번째 숫자(변수)
b = 5           # 두 번째 숫자(변수)

add = a + b     # 덧셈
sub = a - b     # 뺄셈
mul = a * b     # 곱셈
div = a / b     # 나눗셈

print("a =", a)                 # 변수 값 출력
print("b =", b)                 # 변수 값 출력
print("a + b =", add)           # 덧셈 결과 출력
print("a - b =", sub)           # 뺄셈 결과 출력
print("a * b =", mul)           # 곱셈 결과 출력
print("a / b =", div)           # 나눗셈 결과 출력
```
