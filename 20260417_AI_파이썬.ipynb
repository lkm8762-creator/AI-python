```python
prices = {'아메리카노': 2000, '라떼': 3000, '녹차': 2500}
# 1. keys() 활용: 이름표만 확인
for key in prices.keys():
 print(key, end=' ') # 아메리카노 라떼 녹차
print()
# 2. values() 활용: 가격들만 확인
for val in prices.values():
 print(val) # 2000 \n 3000 \n 2500
print()
# 3. items() 활용: 이름과 가격을 동시에!
for key, value in prices.items():
 print(f"{key}의 가격은 {value}원입니다.")
```

```python
#주문 받고 총 금액 계산하기(음료 종류와 수량을 각각 입력받고 곱하기, 반복시키고 더하기, 종료어, 영수증은 order를 누계해야하니 따로 리스트를 만들어서 추가해둬야할듯)-예외처리도 가능한데 예외처리는 못하겠음
cafe_menu = {"아메리카노": 4500,"카페라떼": 5200,"카푸치노": 5200,"바닐라라떼": 5800,"카페모카": 6000,"카라멜 마키아토": 6100,"콜드브루": 5500,"말차라떼": 6200,"초콜릿라떼": 5800,"레몬에이드": 5900}
total = 0
price = 0
while True:  #주문이 언제까지 반복될지 모르니 while이 적합
  order=str(input('메뉴명을 입력해주세요.(정지입력시 종료)'))  #문자로 입력 받고 그걸 키 값으로 딕셔너리 값을 불러냄
  if order == '정지':
    break
  if order in cafe_menu:
    count = int(input('수량을 입력해주세요.'))
    total+= cafe_menu[order]*count
  else:   #메뉴명 예외처리
    print('해당하는 메뉴가 없습니다.')
    continue
print(f'총 금액은 {total}입니다.')
```

```python
#주문 받고 총 금액 계산하기(음료 종류와 수량을 각각 입력받고 곱하기, 반복시키고 더하기, 종료어, 영수증은 order를 누계해야하니 따로 리스트를 만들어서 추가해둬야할듯)-예외처리도 가능한데 예외처리는 못하겠음
cafe_menu = {"아메리카노": 4500,"카페라떼": 5200,"카푸치노": 5200,"바닐라라떼": 5800,"카페모카": 6000,"카라멜 마키아토": 6100,"콜드브루": 5500,"말차라떼": 6200,"초콜릿라떼": 5800,"레몬에이드": 5900}
total = 0
price = 0
total_order = []
while True:  #주문이 언제까지 반복될지 모르니 while이 적합
  order=str(input('메뉴명을 입력해주세요.(정지입력시 종료)'))  #문자로 입력 받고 그걸 키 값으로 딕셔너리 값을 불러냄
  if order == '정지':
    break
  if order in cafe_menu:
    total_order.append(order)
    count = int(input('수량을 입력해주세요.'))
    total+= cafe_menu[order]*count
  else:   #메뉴명 예외처리
    print('해당하는 메뉴가 없습니다.')
    continue
print(f'주문 메뉴는 {total_order}, 총 금액은 {total}입니다.')
```
