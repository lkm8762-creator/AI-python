```python
#변수 제한 사항 등 정
#변수 쓸땐 숫자 우선 불가(ex.2age), 공백 띄워쓰기 불가(무조건 _로 ), 특수문자 대부분 불가, 대소문자 구별은 ㄱㄴ, 파이썬이 이미 문법에 쓰는 단어(예약어)는 변수로 지정 불가
# 변수는 식별자로서 중복된게 있으면 나중에 들어간 값이 출력됨
#변수 하나당 단 하나의 값만 넣을 수 있기에 한 변수에 값을 두 개 주면 앞서 입력된 값은 사라지고 이후에 대입된 값만 남게 됨.
#첫 글자 제한: 반드시 문자 또는 밑줄( _ )로 시작해야 합니다.
#✓ age , _name
#X 1st_age (숫자로 시작할 수 없음)
#구성 요소: 문자, 숫자, 밑줄( _ )만 사용할 수 있습니다. (공백이나 특수문자 불가)
#대소문자 구분: 파이썬은 대소문자를 엄격히 구분합니다. Age 와 age 는 완전히 다른 상자입니다.
#키워드(Keyword) 사용 불가: 파이썬 언어 자체에서 이미 용도를 정해둔 단어(예약어)는 변수 이름으로 쓸 수 없습니다.
```

```python
a=3
b='apple'
c=3.141592
print(type(a)) #정수형
print(type(b)) #문자형, 문자형은 프린트던 대입이던 ''나 ""가 꼭 붙어야
print(type(c)) #소수점, 분수 등 실수형
#그럼 print(type(343, 'ㄹㄴㄹㅇㄴ')) 이런식으로 두 개가 들어가면?

print('hi!'+'My name is...') # +를 붙이면 문자도 이어붙일 수 있음, '',''로 잇는 것도 ㄱㄴ
print('hi!','My name is...')
```

```python
#문제: 분을 시간으로 환산해주는 프로그램
min=int(input('분을 입력해주세요.'))
hour=float(min//60)  #시간 변환
min2=float(min%60)  #분 변환
print(min,'분은', hour,'시간',min2,'분 입니다.')
```

```python
n=2
n+=4
m=n+4   #혼합연산자 이해가 잘안됨
print(n, m)
```

```python
num = int(input("정수를 입력하세요: "))

# num을 2로 나눈 나머지가 0이면 짝수
if num % 2 == 0:
    print("짝수입니다.")
else:
    print("홀수입니다.")
```

```python
num = int(input("정수를 입력하세요: "))

# 정수는 양수, 0, 음수로 구분됨. 0보다 크거나 작거나 같거나로 구분
if num >0:
    print("양수입니다.")
elif num==0:
    print('0입니다.')
else:
    print("음수입니다.")
```

```python
#문제: 4가지 메뉴에 대해 각각의 주문 수와 총 가격을 계산하기
#메뉴판
#coffee=2000
#latte=3000
#greentea=2500
#juice=2800
coffee=int(input('커피 개수를 입력해주세요:'))
latte=int(input('라떼 개수를 입력해주세요:'))
greentea=int(input('녹차 개수를 입력해주세요:'))
juice=int(input('주스 개수를 입력해주세요:'))
total_price=(2000*coffee)+(3000*latte)+(2500*greentea)+(2800*juice)
print('커피',coffee,'잔', '라떼',latte,'잔', '녹차',greentea,'잔', '주스',juice,'잔', '하여 주문하신 음료의 총 가격은',total_price,'원입니다.')
```
